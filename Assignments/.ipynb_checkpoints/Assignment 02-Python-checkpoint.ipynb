# 3.2
## Carefully explain the diferences between the KNN classifer and KNN regression methods.

K-Nearest Neighbors (KNN) is a versatile algorithm that can be used for both classification and regression tasks. The main difference between KNN classifier and KNN regression lies in how they predict the output for new data points.

### KNN Classifier:
In KNN classification, the algorithm predicts the class membership of a data point based on the class labels of its nearest neighbors. Here's how it works:

1. **Distance Calculation**: For a new data point, the algorithm calculates the distances to all other data points in the training set. Common distance metrics include Euclidean distance, Manhattan distance, or cosine similarity.
   
2. **Nearest Neighbors Selection**: It selects the K nearest neighbors based on the calculated distances.
   
3. **Voting**: The algorithm assigns the class label to the new data point by taking a majority vote among the labels of its K nearest neighbors. The class with the highest number of occurrences among the neighbors is assigned as the predicted class for the new data point.

### KNN Regression:
In KNN regression, instead of predicting a class label, the algorithm predicts a continuous value for the target variable. Here's how it differs from the classifier:

1. **Distance Calculation**: Similar to classification, the algorithm calculates distances between the new data point and all other data points in the training set.

2. **Nearest Neighbors Selection**: It selects the K nearest neighbors based on the calculated distances.

3. **Averaging**: Instead of taking a majority vote, KNN regression computes the average (or sometimes weighted average) of the target values of the K nearest neighbors. This average value is then assigned as the predicted output for the new data point.

### Summary of Differences:
1. **Output Type**: KNN classifier predicts discrete class labels, while KNN regression predicts continuous values.
2. **Prediction Mechanism**: Classifier uses majority voting for class prediction, whereas regression uses averaging for value prediction.
3. **Evaluation Metrics**: Since they predict different types of outputs, the evaluation metrics used for classifier and regression tasks also differ. For classification, metrics like accuracy, precision, recall, and F1 score are common, while for regression, metrics like mean squared error (MSE) or mean absolute error (MAE) are used.

In summary, while both KNN classifier and KNN regression rely on the same basic principle of finding nearest neighbors, they differ in how they utilize this information to make predictions based on the nature of the output variable.

### (c) Use the sm.OLS() function to perform a multiple linear regression with mpg as the response and all other variables except name as the predictors. Use the summarize() function to print the results. Comment on the output. For instance:
    

```python
# Define predictors (all variables except 'name') and response variable
predictors = auto.drop(columns=['name', 'mpg'])  # Exclude 'name' and 'mpg'
response = auto['mpg']

# Add constant term for intercept
predictors_with_const = sm.add_constant(predictors)

# Fit multiple linear regression model
model = sm.OLS(response, predictors_with_const)
results = model.fit()

# Define significance level
significance_level = 0.05

# Print summary of results
print(results.summary())
```

#### i. Is there a relationship between the predictors and the response? Use the anova_lm() function from statsmodels to answer this question.

```python
# i. Check if there is a significant relationship between predictors and the response variable
if results.f_pvalue < significance_level:
    print("i. There is a significant relationship between predictors and the response variable.")
else:
    print("i. There is no significant relationship between predictors and the response variable.")

```

#### ii. Which predictors appear to have a statistically signifcant relationship to the response?

```python
# ii. Identify predictors with statistically significant relationships to the response variable
significant_predictors = results.pvalues[results.pvalues < significance_level].index.tolist()
print("ii. Predictors with statistically significant relationships to the response variable:", significant_predictors)
```

#### iii. What does the coefcient for the year variable suggest?

```python
# iii. Interpret the coefficient for the 'year' variable
year_coef = results.params['year']
if year_coef > 0:
    print("iii. The coefficient for the 'year' variable is positive, suggesting an increase in mpg as the year increases, indicating a potential improvement in fuel efficiency over time.")
else:
    print("iii. The coefficient for the 'year' variable is negative, which might be unexpected and needs further investigation.")
```

### Try a few diferent transformations of the variables, such as log(X), X, X2. Comment on your fndings.

```python
# Logarithmic transformation
predictors_log = predictors.copy()
predictors_log['log_displacement'] = np.log(predictors_log['displacement'])
predictors_log['log_horsepower'] = np.log(predictors_log['horsepower'])
predictors_log['log_weight'] = np.log(predictors_log['weight'])

# Fit multiple linear regression model with logarithmic transformation
model_log = sm.OLS(response, sm.add_constant(predictors_log))
results_log = model_log.fit()

# Print summary of results for logarithmic transformation
print("Logarithmic Transformation:")
print(results_log.summary())
print()

# Square root transformation
predictors_sqrt = predictors.copy()
predictors_sqrt['sqrt_displacement'] = np.sqrt(predictors_sqrt['displacement'])
predictors_sqrt['sqrt_horsepower'] = np.sqrt(predictors_sqrt['horsepower'])
predictors_sqrt['sqrt_weight'] = np.sqrt(predictors_sqrt['weight'])

# Fit multiple linear regression model with square root transformation
model_sqrt = sm.OLS(response, sm.add_constant(predictors_sqrt))
results_sqrt = model_sqrt.fit()

# Print summary of results for square root transformation
print("Square Root Transformation:")
print(results_sqrt.summary())
print()

# Squared transformation
predictors_squared = predictors.copy()
predictors_squared['displacement_squared'] = predictors_squared['displacement'] ** 2
predictors_squared['horsepower_squared'] = predictors_squared['horsepower'] ** 2
predictors_squared['weight_squared'] = predictors_squared['weight'] ** 2

# Fit multiple linear regression model with squared transformation
model_squared = sm.OLS(response, sm.add_constant(predictors_squared))
results_squared = model_squared.fit()

# Print summary of results for squared transformation
print("Squared Transformation:")
print(results_squared.summary())
```

```python
# Define a function to extract information for each model
def extract_model_info(results, name):
    model_info = {
        "R-squared": results.rsquared,
        "Intercept": results.params['const'],
    }
    # Extract information for each predictor variable
    for predictor in predictors.columns:
        model_info[f"{predictor} Coefficient"] = results.params[predictor]
        model_info[f"{predictor} p-value"] = results.pvalues[predictor]
    return {name: model_info}

# Extract relevant information from each model
models_info = {}

# Original Model
models_info.update(extract_model_info(results, "Original Model"))

# Logarithmic Transformation
models_info.update(extract_model_info(results_log, "Logarithmic Transformation"))

# Square Root Transformation
models_info.update(extract_model_info(results_sqrt, "Square Root Transformation"))

# Squared Transformation
models_info.update(extract_model_info(results_squared, "Squared Transformation"))

# Create a DataFrame from the dictionary
models_df = pd.DataFrame(models_info).T

# Display the DataFrame
print(models_df)
```

#### R-squared value: Higher R-squared values indicate better fit of the model to the data. Therefore, preference on the model with the highest R-squared value.


```python
# Find the model with the highest R-squared value
best_model_r_squared = models_df.loc[models_df['R-squared'].idxmax()]
print("Best Model based on R-squared value:")
print(best_model_r_squared)
```

#### Statistical significance of predictors: Check the p-values associated with each predictor in the model. Statistically significant predictors (p-value < 0.05) are more likely to have a true relationship with the response variable.

```python
# Initialize variables to keep track of the best model
best_model = None
best_significant_predictors_count = 0

# Iterate over each model
for model_name, model_info in models_info.items():
    # Count the number of statistically significant predictors (p-value < 0.05)
    significant_predictors_count = sum(1 for info in model_info.values() if isinstance(info, float) and info < 0.05)
    
    # Update the best model if the current model has more significant predictors
    if significant_predictors_count > best_significant_predictors_count:
        best_model = model_name
        best_significant_predictors_count = significant_predictors_count

# Print the best model and the number of significant predictors
if best_model:
    print("Best Model:", best_model)
    print("Number of Significant Predictors:", best_significant_predictors_count)
else:
    print("No model has statistically significant predictors.")
```

## (a) 
### Fit a multiple regression model to predict Sales using Price, Urban, and US.


```python
# Define the predictors and response variable
X = carseats[['Price', 'Urban', 'US']]
y = carseats['Sales']

# Convert categorical variables to dummy variables
X = pd.get_dummies(X, drop_first=True)

# Add constant term to the predictors
X = sm.add_constant(X)

# Fit the multiple regression model
modelMR = sm.OLS(y, X).fit()

# Print the summary of the model
print(modelMR.summary())
```

## (b) 
### Provide an interpretation of each coefficient in the model. Be carefulsome of the variables in the model are qualitative!


1. **const (Intercept)**: The intercept represents the expected value of Sales when all predictor variables are zero. In this case, when Price, Urban, and US are all zero, the expected Sales is 13.0435.

2. **Price**: The coefficient of Price (-0.0545) indicates the change in Sales for a one-unit increase in Price, holding all other variables constant. Since the coefficient is negative, it suggests that as the price of the product increases by one unit, Sales are expected to decrease by approximately 0.0545 units.

3. **Urban_Yes**: This coefficient (-0.0219) is associated with the Urban predictor. Since Urban is a dummy variable (1 for 'Yes' and 0 for 'No'), the coefficient represents the difference in Sales between urban and non-urban locations. However, the p-value (0.936) is not significant, indicating that Urban does not have a statistically significant effect on Sales in this model.

4. **US_Yes**: This coefficient (1.2006) is associated with the US predictor. Similar to Urban, it represents the difference in Sales between products sold in the US and those sold elsewhere. The positive coefficient suggests that products sold in the US have higher Sales compared to those sold elsewhere. The p-value (0.000) indicates that this difference is statistically significant.

In summary, Price and whether the product is sold in the US have statistically significant effects on Sales, while the location being urban or not does not have a significant effect.

## (c)
### Write out the model in equation form, being careful to handle the qualitative variables properly.


```python
# Extract coefficients from the model
coefficients = modelMR.params

# Manually format coefficients into a mathematical expression
mathematical_expression = f"Sales = {coefficients['const']:.2f}"
for var, coef in coefficients[1:].items():
    if coef >= 0:
        mathematical_expression += f" + {coef:.2f} * {var}"
    else:
        mathematical_expression += f" - {-coef:.2f} * {var}"

mathematical_expression += " + "  # Adding the epsilon symbol

# Print coefficients and mathematical expression
print(coefficients)
print(mathematical_expression)
```

```python
# Extract coefficients from the model
coefficients = modelMR.params

# Manually format coefficients into a LaTeX expression
latex_expression = "$$ Sales = {:.2f}".format(coefficients['const'])
for var, coef in coefficients[1:].items():
    subscripted_var = var.replace('_Yes', '_{\mathrm{Yes}}')
    if coef >= 0:
        latex_expression += " + {:.2f} \\times {}".format(coef, subscripted_var)
    else:
        latex_expression += " - {:.2f} \\times {}".format(-coef, subscripted_var)

latex_expression += " + \\epsilon $$"

# Print LaTeX expression
from IPython.display import display, Markdown
display(Markdown(latex_expression))
```

### Is there evidence of outliers or high leverage observations in the model from (e)?


```python
# Calculate residuals
residuals = smaller_model.resid

# Calculate leverage statistics
leverage = smaller_model.get_influence().hat_matrix_diag

# Calculate standardized residuals
standardized_residuals = smaller_model.get_influence().resid_studentized_internal

# Find observations with large residuals (potential outliers)
outliers = np.abs(standardized_residuals) > 2  # Using a threshold of 2 for identifying outliers

# Find observations with high leverage
high_leverage = leverage > (2 * (smaller_model.df_model + 1) / smaller_model.nobs)  # Using a threshold of 2 * (p + 1) / n

# Print the indices of potential outliers
print("Indices of potential outliers:", np.where(outliers)[0])

# Print the indices of observations with high leverage
print("Indices of observations with high leverage:", np.where(high_leverage)[0])
```

## This problem involves simple linear regression without an intercept.

## (a) 
### Recall that the coefcient estimate  for the linear regression of Y onto X without an intercept is given by (3.38). Under what circumstance is the coefcient estimate for the regression of X onto Y the same as the coefcient estimate for the regression of Y onto X?


The coefficient estimate for the regression of \( X \) onto \( Y \) will be the same as the coefficient estimate for the regression of \( Y \) onto \( X \) if and only if the product of the standard deviations of \( X \) and \( Y \) is equal to the sample correlation coefficient between \( X \) and \( Y \).

Mathematically, this condition can be expressed as:

$
\frac{{\text{{cov}}(X, Y)}}{{\text{{Var}}(X)}} = \frac{{\text{{cov}}(Y, X)}}{{\text{{Var}}(Y)}}
$

This condition essentially ensures that the regression lines will have the same slope when fitting \( X \) as the response variable or when fitting \( Y \) as the response variable. It's important to note that this condition holds true only for simple linear regression without an intercept.

## (b) 
### Generate an example in Python with n = 100 observations in which the coefficient estimate for the regression of X onto Y is different from the coefficient estimate for the regression of Y onto X.


```python
# Set seed for reproducibility
np.random.seed(123)

# Generate data
n = 100
X_diff = np.random.normal(loc=0, scale=1, size=n)
Y_diff = 2 * X_diff + np.random.normal(loc=0, scale=1, size=n)  # Corrected line

# Regression of X onto Y
model_X_on_Y_diff = sm.OLS(X_diff, sm.add_constant(Y_diff)).fit()  # Corrected line

# Regression of Y onto X
model_Y_on_X_diff = sm.OLS(Y_diff, sm.add_constant(X_diff)).fit()  # Corrected line

# Print coefficient estimates
print("Coefficient estimate for X onto Y:", model_X_on_Y_diff.params[1])
print("Coefficient estimate for Y onto X:", model_Y_on_X_diff.params[1])

# Plot data points and regression lines
plt.scatter(Y_diff, X_diff, label='Data')
plt.plot(Y_diff, model_X_on_Y_diff.predict(sm.add_constant(Y_diff)), color='red', label='X onto Y')
plt.plot(X_diff, model_Y_on_X_diff.predict(sm.add_constant(X_diff)), color='green', label='Y onto X')  # Corrected line
plt.xlabel('Y')
plt.ylabel('X')
plt.title('Regression of X onto Y and Y onto X (Different Coefficients)')
plt.legend()
plt.show()
```

## (c) 
### Generate an example in Python with n = 100 observations in which the coefficient estimate for the regression of X onto Y is the same as the coefficient estimate for the regression of Y onto X.

```python
# Set seed for reproducibility
np.random.seed(456)

# Generate data
n = 100
X_same = np.random.normal(loc=0, scale=1, size=n)
Y_same = X_same 

# Regression of X onto Y
model_X_on_Y_same = sm.OLS(X_same, sm.add_constant(Y_same)).fit()

# Regression of Y onto X
model_Y_on_X_same = sm.OLS(Y_same, sm.add_constant(X_same)).fit()

# Check if the coefficient estimates for X onto Y and Y onto X are the same
if model_X_on_Y_same.params[1] == model_Y_on_X_same.params[1]:
    print("The slopes are the same.")
else:
    print("The slopes are different.")

# Print coefficient estimates
print("Coefficient estimate for X onto Y:", model_X_on_Y_same.params[1])
print("Coefficient estimate for Y onto X:", model_Y_on_X_same.params[1])

# Plot data points and regression lines
plt.scatter(X_same, Y_same, label='Data')
plt.plot(X_same, model_X_on_Y_same.predict(sm.add_constant(X_same)), color='red', label='X onto Y')
plt.plot(X_same, model_Y_on_X_same.predict(sm.add_constant(X_same)), color='green', label='Y onto X')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression of X onto Y and Y onto X (Same Coefficients)')
plt.legend()
plt.show()
```
