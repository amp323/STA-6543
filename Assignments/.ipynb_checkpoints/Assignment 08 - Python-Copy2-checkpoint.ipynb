### (e) Now fit a logistic regression model to the data using non-linear functions of $X_1$ and $X_2$ as predictors (e.g. $X_{12}$, $X_1 \times X_2$, $log(X_2)$, and so forth).


```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Define feature matrix X
X = np.column_stack((x1, x2))

# Define different transformations
transformations = [
    ('X1*X2', lambda X: X[:, 0] * X[:, 1]),                 # X1 * X2
    ('X1^2', lambda X: X[:, 0] ** 2),                       # X1^2
    ('X2^2', lambda X: X[:, 1] ** 2),                       # X2^2
    ('log(X1)', lambda X: np.log(X[:, 0] + 1e-5)),          # log(X1)
    ('log(X2)', lambda X: np.log(X[:, 1] + 1e-5)),          # log(X2)
]

best_accuracy = 0
best_model = None

# Iterate over each transformation
for name, transform in transformations:
    # Apply transformation to features
    X_transformed = transform(X)
    X_transformed = X_transformed.reshape(-1, 1)  # Reshape to 2D array
    
    # Check for missing values
    if np.isnan(X_transformed).any() or np.isinf(X_transformed).any():
        print(f"Transformation {name} resulted in NaN or Inf values. Skipping...")
        continue
    
    # Fit logistic regression model using cross-validation
    logistic_model = LogisticRegression()
    scores = cross_val_score(logistic_model, X_transformed, y, cv=5)
    accuracy = scores.mean()
    
    # Update best model if accuracy is improved
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = (name, logistic_model)

# Print the best model and its accuracy
print("Best model:", best_model[0])
print("Accuracy:", best_accuracy)

# Save the best model to a variable for later use
best_transform, best_logistic_model = best_model
print("Best model:", best_model)
```
